# กรองสถานี

```python
#กรองข้อมูลให้เหลือเฉพาะสถานีใน stations in the model_Hill.csv แสดงจำนวนสถานีก่อน/หลังกรอง
before = daily_2025['tele_station_oldcode'].nunique()

daily_2025 = daily_2025[
    daily_2025['tele_station_oldcode'].astype(str).isin(stations['Node'].astype(str))
]

after = daily_2025['tele_station_oldcode'].nunique()

print(f"จำนวนสถานีก่อนกรอง: {before}")
print(f"จำนวนสถานีหลังกรอง: {after}")
```

มีสถานี 384 แห่งจากทั้งหมด 1514 สถานีที่ตรงกับในไฟล์ stations in the model_Hill.csv

# record ที่หายไป

```python
#ระบุข้อมูลสูญหาย (record ที่หายไป) ของแต่ละสถานี แสดงจำนวนและสัดส่วน
# Make sure 'measure_datetime' is in datetime format
daily_2025 = daily_2025.copy() # Add .copy() to prevent SettingWithCopyWarning
daily_2025['measure_datetime'] = pd.to_datetime(daily_2025['measure_datetime'])

# Get all unique stations
all_stations = daily_2025['tele_station_oldcode'].unique()

# Determine the overall min and max date in the dataset
min_date = daily_2025['measure_datetime'].min()
max_date = daily_2025['measure_datetime'].max()

missing_data_info = []

for station_code in all_stations:
    # Filter data for the current station
    station_df = daily_2025[daily_2025['tele_station_oldcode'] == station_code]

    # Create a complete date range for the station
    full_date_range = pd.date_range(start=min_date, end=max_date, freq='D')
    full_dates_df = pd.DataFrame({'measure_datetime': full_date_range})

    # Merge to find missing dates (records)
    merged_df = pd.merge(full_dates_df, station_df, on='measure_datetime', how='left')

    # Count missing records (where 'tele_station_id' is NaN after merge)
    missing_records_count = merged_df['tele_station_id'].isnull().sum()
    total_possible_records = len(full_date_range)
    missing_proportion = missing_records_count / total_possible_records if total_possible_records > 0 else 0

    missing_data_info.append({
        'Station': station_code,
        'Missing Records Count': missing_records_count,
        'Missing Records Proportion': missing_proportion
    })

missing_data_df = pd.DataFrame(missing_data_info)

display(missing_data_df.head())
print(f"\nNumber of stations with missing data: {len(missing_data_df[missing_data_df['Missing Records Count'] > 0])}")
```

# คัดสถานีที่มีข้อมูลสูญหาย >=20% ออก

```python
#คัดสถานีที่มีข้อมูลสูญหาย >=20% ออก แสดงรายชื่อสถานีที่ถูกคัดออกและจำนวนที่เหลือ
# Identify stations to be removed (missing records >= 20%)
stations_to_remove = missing_data_df[missing_data_df['Missing Records Proportion'] >= 0.20]

# Get the list of station codes to remove
removed_station_codes = stations_to_remove['Station'].tolist()

# Display the list of removed stations and their count
print("รายชื่อสถานีที่ถูกคัดออกเนื่องจากมีข้อมูลสูญหาย >= 20%:")
if len(removed_station_codes) > 0:
    for station in removed_station_codes:
        print(f"- {station}")
else:
    print("ไม่มีสถานีใดถูกคัดออกในเกณฑ์นี้")

print(f"\nจำนวนสถานีที่ถูกคัดออก: {len(removed_station_codes)}")

# Filter daily_2025 to keep only stations not in the removed list
before_removal_count = daily_2025['tele_station_oldcode'].nunique()
daily_2025 = daily_2025[~daily_2025['tele_station_oldcode'].isin(removed_station_codes)]
after_removal_count = daily_2025['tele_station_oldcode'].nunique()

print(f"จำนวนสถานีก่อนคัดออก: {before_removal_count}")
print(f"จำนวนสถานีที่เหลือหลังจากคัดออก: {after_removal_count}")
```

# เติมข้อมูลที่หายไปด้วยค่า Median

```python
# Ensure 'measure_datetime' in historical is datetime
historical['measure_datetime'] = pd.to_datetime(historical['measure_datetime'])

# Extract Day of Year from historical data
historical['doy'] = historical['measure_datetime'].dt.dayofyear

# Calculate median rainfall for each station and day of year from historical data
median_table = historical.groupby(['station_code', 'doy'])['data'].median().reset_index()

# -----------------------------------------------------------------------------
# Revised logic to handle missing records (entire dates)
# -----------------------------------------------------------------------------

# Get all unique stations that remain after filtering
unique_stations = daily_2025['tele_station_oldcode'].unique()

# Determine the overall min and max date in the daily_2025 dataset
min_date_daily = daily_2025['measure_datetime'].min()
max_date_daily = daily_2025['measure_datetime'].max()

full_date_range = pd.date_range(start=min_date_daily, end=max_date_daily, freq='D')

# Create a DataFrame with all possible (station, date) combinations
all_possible_records = pd.MultiIndex.from_product(
    [unique_stations, full_date_range],
    names=['tele_station_oldcode', 'measure_datetime']
).to_frame(index=False)

# Merge with the existing daily_2025 data to fill in actual observations
daily_2025_full = pd.merge(
    all_possible_records,
    daily_2025.drop(columns=['doy'], errors='ignore'), # Drop doy if exists from previous run before re-adding
    on=['tele_station_oldcode', 'measure_datetime'],
    how='left'
)

# Extract Day of Year for the full dataset
daily_2025_full['doy'] = daily_2025_full['measure_datetime'].dt.dayofyear

# Merge with median_table to bring in median rainfall values for imputation
daily_2025_imputed = pd.merge(
    daily_2025_full,
    median_table,
    left_on=['tele_station_oldcode', 'doy'],
    right_on=['station_code', 'doy'],
    how='left',
    suffixes=('', '_median')
)

# Fill missing 'rainfall_value' with the median from 'data' column
daily_2025_imputed['rainfall_value'] = daily_2025_imputed['rainfall_value'].fillna(daily_2025_imputed['data'])

daily_2025_imputed.drop(columns=['doy', 'station_code', 'data'], inplace=True)

display(daily_2025_imputed.head())
```

```python
daily_2025_imputed['value'] = (
    daily_2025_imputed['value']
    .fillna(daily_2025_imputed['station_code_median'])
)
```

จากการตรวจสอบค่า missing หลังการเติมข้อมูล พบว่า**คอลัมน์ที่ใช้ในการวิเคราะห์** ได้แก่ วันที่และค่าปริมาณน้ำฝน **ไม่มีค่า NaN** เหลืออยู่ แสดงว่าข้อมูลรายวันมีความสมบูรณ์และพร้อมสำหรับการคำนวณผลรวมรายเดือน

## ใช้ Median แทน Mean เพราะข้อมูลปริมาณน้ำฝนมีลักษณะการกระจายตัวแบบเบ้ขวา และ Median สามารถทนต่อผลกระทบของ Outliers ได้ดีกว่า Mean

# คำนวณผลรวม

```python
# แปลงคอลัมน์วันที่ให้เป็น datetime
daily_2025_imputed['measure_datetime'] = pd.to_datetime(
    daily_2025_imputed['measure_datetime']
)

# สร้าง year และ month
daily_2025_imputed['year'] = daily_2025_imputed['measure_datetime'].dt.year
daily_2025_imputed['month'] = daily_2025_imputed['measure_datetime'].dt.month

# รวมปริมาณน้ำฝนรายเดือน
monthly_rain = (
    daily_2025_imputed
    .groupby(['tele_station_oldcode', 'year', 'month'])['rainfall_value']
    .sum()
    .reset_index()
)

# เปลี่ยนชื่อคอลัมน์ให้ตรงกับไฟล์ตัวอย่าง
monthly_rain = monthly_rain.rename(columns={
    'tele_station_oldcode': 'Node',
    'rainfall_value': 'rain'
})

monthly_rain = monthly_rain.sort_values(
    by=['Node', 'year', 'month']
)

monthly_rain = monthly_rain[['Node', 'rain', 'month', 'year']]

monthly_rain.head()
```
