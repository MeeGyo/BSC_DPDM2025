# นักแสดงคนไหนที่มักจะพบว่าเล่นหนังด้วยกัน

```python
df = India_data_cleaned
```

จัดกลุ่ม Rating ออกเป็น4กลุ่ม โดยอิงจากการกระจายของข้อมูลตามควอไทล์ เพื่อวิเคราะห์ระดับคุณภาพของข้อมูล

```python
df['Rating_Quantiles'] = pd.qcut(
    df['Rating'],
    q=4,
    labels=['Poor', 'Average', 'Good', 'Excellent']
)
```

แปลงคอลัมน์นักแสดงให้แยกจากกัน

```python
actors = pd.get_dummies(
    df[['Actor 1','Actor 2','Actor 3']].stack(),
    prefix='Actor'
).groupby(level=0).max()
```

```python
rating_dummies = pd.get_dummies(
    df['Rating_Quantiles'],
    prefix='Rating_Quantiles'
)
```

```python
data_apri = pd.concat([actors, rating_dummies], axis=1)
data_apri.head()
```

เลือกเมื่อปรากฏในภาพยนตร์อย่างน้อย 0.1% ของจำนวนภาพยนตร์ทั้งหมด

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(
    data_apri,
    min_support=0.001,
    use_colnames=True
)
```

```python
rules1 = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
)

rules1.sort_values('lift', ascending=False).head(10)
```

# นักแสดงคนไหนเล่นหนังประเภทไหน ได้rating Good/Excellent

ตรวจสอบว่านักแสดงคนไหนเล่นไหนประเภทไหน ได้เรตติ้งเท่าไหร่



```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

df2 = India_data_cleaned[
    ['Actor 1', 'Actor 2', 'Actor 3', 'Genre', 'Rating_Quantiles']
].copy()

actors_long = df2.melt(
    id_vars=['Genre', 'Rating_Quantiles'],
    value_vars=['Actor 1', 'Actor 2', 'Actor 3'],
    value_name='Actor'
)

actors_long = actors_long.dropna(subset=['Actor'])

actors_long['Genre'] = actors_long['Genre'].str.split(',')

actor_genre_long = actors_long.explode('Genre')
actor_genre_long['Genre'] = actor_genre_long['Genre'].str.strip()

basket = actor_genre_long[['Actor', 'Genre', 'Rating_Quantiles']].copy()

basket['Actor'] = 'Actor_' + basket['Actor'].astype(str)
basket['Genre'] = 'Genre_' + basket['Genre'].astype(str)
basket['Rating_Quantiles'] = 'Rating_' + basket['Rating_Quantiles'].astype(str)

transactions = basket.apply(list, axis=1)

te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df_te = pd.DataFrame(te_array, columns=te.columns_)
```

```python
frequent_itemsets = apriori(
    df_te,
    min_support=0.00065,
    use_colnames=True,
    max_len=3   # Actor + Genre + Rating
)
```

```python
rules = association_rules(
    frequent_itemsets,
    metric='confidence',
    min_threshold=0.3
)
```

```python
actor_genre_rating_rules = rules[
    rules['antecedents'].apply(
        lambda x: any(i.startswith('Actor_') for i in x)
        and any(i.startswith('Genre_') for i in x)
    ) &
    rules['consequents'].apply(
        lambda x: any(i.startswith('Rating_') for i in x)
    )
].sort_values(['lift', 'confidence'], ascending=False)
```

```python
actor_genre_rating_rules.head(10)
```

```python
actor_genre_rating_rules.shape
```

ดูเฉพาะ Rating Good/Excellent

```python
good_ratings = ['Rating_Good', 'Rating_Excellent']

actor_genre_good_rules = rules[
    rules['antecedents'].apply(
        lambda x: any(i.startswith('Actor_') for i in x)
        and any(i.startswith('Genre_') for i in x)
    ) &
    rules['consequents'].apply(
        lambda x: any(i in good_ratings for i in x)
    )
].sort_values(['lift', 'confidence'], ascending=False)
```
